```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/C45Nishit/AICS-LAB/main/Social_Network_Ads.csv")
x=df['Age'].values
y=df['EstimatedSalary'].values
c=df['Purchased'].values
a=[]
for i in range(len(x)):
  a.append([x[i],y[i]])
plt.scatter(x,y,c=c)
```

```python
k=KNeighborsClassifier()
k.fit(a,c)
k.predict([(int(input("age:")),int(input("salary:")))])[0]
```
